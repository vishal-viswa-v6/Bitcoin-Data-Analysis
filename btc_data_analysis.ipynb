## Bitcoin Price Data Analysis, January 1st 2023 to June 4th 2025, Vishal Viswa

```python
import pandas as pd
import numpy as np
import yfinance as yf 
import matplotlib.pyplot as plt
```

```python
data = yf.download('BTC-USD', period='5y')
```

```python
data.head()
```

```python
data.info()
```

```python
print(data.columns)
```

```python
data.columns = data.columns.get_level_values(0)
data['Daily Return'] = data['Close'].pct_change()
```

### Plotting Closing Price

```python
plt.figure(figsize=(12, 5))
plt.plot(data['Close'], label = 'BTC-USD Closing Price')
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Plotting Daily Returns

```python
plt.figure(figsize = (12, 4))
plt.plot(data['Daily Return'], color = 'orange', label = 'Daily Return')
plt.title('Bitcoin Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.axhline(0, color='gray', linestyle = '--', linewidth = 1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
